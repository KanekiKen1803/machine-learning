```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = './data_1.xlsx'
q1_a = pd.read_excel(filepath)
```

```python
q1_a.head()
```

```python
plt.figure(figsize=(10,5))
plt.title("Scatter Plot")
sns.scatterplot(data = q1_a, x = q1_a['x'], y = q1_a['y'])
```

```python
plt.title("Histogram for X")
sns.histplot(data = q1_a, x = q1_a['x'])

plt.figure()
plt.title("Histogram for Y")
sns.histplot(data = q1_a, x = q1_a['y'])
plt.figure()

plt.title("Histogram Combined")
sns.histplot(data = q1_a)
```

```python
plt.figure(figsize=(5,10))
sns.heatmap(q1_a)
plt.title("Conventional heatmap with two features")
plt.ylabel("Values")

plt.figure()
sns.displot(q1_a, x=q1_a['x'], y = q1_a['y'], cbar = True)
plt.title("Density Heatmap")
```

```python
plt.figure(figsize=(5,10))
sns.boxplot(data =q1_a)
plt.title("BoxPlot for x and y")
```
